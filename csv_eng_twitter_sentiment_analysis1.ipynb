```python
import io 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
import nltk
nltk.download('stopwords')
stop_words = set(stopwords.words('english')) 
```

```python
import re
def process(data):
    data = data.lower()
    data = re.sub(".[!@#$+%*:()'-]", ' ', data) # simbol
    data = re.sub(r'\d+', '', data) # angka
    data = re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", data) # all alphabet
    data = data.replace("   ", " ")
    data = data.replace("  ", " ")
    data = data.replace("\n","") #hilangin \n atau pake .stripe() /rstripe()
    data = re.sub('((http://)[^ ]*|(https://)[^ ]*|( www\.)[^ ]*)', 'URL', data) # remove URLs
    data = re.sub('@[^\s]+', ' USER', data) # remove usernames
    data = re.sub(r'#([^\s]+)', r'\1', data) # remove the # in #hashtag
    data = word_tokenize(data) # remove repeated characters (helloooooooo into hello)
    return [word for word in data if word not in stop_words] # kembali ke w, untuk w pada text jika w tidak pada list data twitter yang sudah dibersihkan
  
```

```python
import time
t = time.time()
poin, sentiment = df['content'], df['sentiment']
poin = poin.drop_duplicates()
print(poin)
content=poin.apply(process)
#print(f'Text Preprocessing complete.')
#print(f'Time Taken: {round(time.time()-t)} seconds')
```
